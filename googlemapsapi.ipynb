```python
#Imports

import pandas as pd
import googlemaps
from datetime import datetime

gmaps = googlemaps.Client(key='')
```

```python
#read csv to transform to dataframe 
df = pd.read_csv('Stations_Data.csv')
df.head()

df2 = pd.read_excel('Barris.xlsx')
df2['PostalCode'] = df2['PostalCode'].str.replace("'","")

df2.head()
```

```python
#Replace character to make intersection 
df['Name'] = df['Name'].str.replace('|','&')

# Find zicode from streets

for index, row in df.iterrows():
    geocode_result = gmaps.geocode( str(row['Name']) + ',BARCELONA , ES')
    print(geocode_result)
    a= geocode_result[0]['address_components']
    size =len(a) -1
    posta_code = (a[size]['long_name'])
    df.at[index,"Postalcode"] =posta_code
```

```python
#Find hood in df2 
for index, row in df.iterrows():
    postalcode = row['Postalcode']
    barri =  df2.loc[df2['PostalCode'] == postalcode]
    hood = ''
    for index1, row1 in barri.iterrows():
        hood = row1['Neighbourhood']

    df.at[index,"Barri"] =str(hood)
```

```python
#Extract DataFrame to excel 

df.to_excel("result.xlsx")  
```
